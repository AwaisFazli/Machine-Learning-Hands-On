```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
GDP = pd.read_csv("drive/MyDrive/Book Datasets/gdp_per_capita.csv", sep=",")
BLI = pd.read_csv("drive/MyDrive/Book Datasets/BetterLifeIndex2015.csv")
```

```python
GDP.head()
```

```python
BLI.head()
```

```python
gdp = GDP[["Country", "2015"]]
bli = BLI[["Country", "  Life satisfaction"]]
```

```python
merged = pd.merge(left=gdp, right=bli, left_on='Country', right_on='Country')
```

```python
merged.head()
```

```python
plt.scatter(merged["2015"], merged["  Life satisfaction"])
```

```python
X = merged["2015"].values.reshape(-1,1)
Y = merged["  Life satisfaction"].values.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(X , Y)
```

```python
X_new = [[22587]] #Cyprus GDP per capita
print(model.predict(X_new))
```
